Conclusion:  In 2017, Fort Hamilton High School had a mean score of 71 for the Commom Core Algebra, with 64% of the student who took the test scoring at a College Readiness level.  For that same year and the same test, the overall High School mean was 65, with 43% of the student who took the test scoriing at a College Readiness Level.  So Ft. Hamilton students scored higher than the overall average high school scores.
